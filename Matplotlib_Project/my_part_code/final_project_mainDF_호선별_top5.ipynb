##### 연도별 역별 총합 하차인원 DF
- 1월 -12월 데이터 행 합쳐서 '총합' 칼럼 "년" 옆에 생성

```python
year_group_Df = rawDF.groupby("년")
year_group_Df.get_group(2006)
```

```python
sum_year = dict(zip( rawDF["년"].unique().tolist(), [0]*len(rawDF["년"].unique()) ) ) 

sum_year
```

```python
# sum_year = dict(zip( rawDF["년"].unique().tolist(), [0]*len(rawDF["년"].unique()) ) ) 

for year in rawDF["년"].unique():
    sum_year[year] = year_group_Df.get_group(year).sum()[2:].tolist()

sum_year[2006]
```

```python
# 연도별 역별 하차인원 총합 DF 생성

year_out_DF = pd.DataFrame( sum_year, index=rawDF.columns[2:])
year_out_DF_final = year_out_DF.T.copy()
```

##### 호선 별 2023 -2015 총 하차 이용객 TOP 8 그래프
- 막대 그래프

```python
def station_tot(df_line):
    
    tmp = df_line.loc[2015:2023, :].sum(axis=0).sort_values(ascending=False)

    station = tmp[:8].index.tolist()
    total   = tmp[:8].tolist()

    return station, total
```

```python
# subplots Ver.

# x축, y축 데이터 생성
first_station, first_tot = station_tot(first_line)
second_station,second_tot  = station_tot(second_line)
third_station, third_tot  = station_tot(third_line)


# bar plot
fig = plt.figure( figsize=(20,28) )
axes = fig.subplots(3,1)
station_val=[ [first_station, first_tot], [second_station, second_tot], [third_station, third_tot]]
color_list = ["red", "green", "gold"]
ylim_list = [1,4, -10**(0.7)]

for idx, col in enumerate(axes):
    bar1 = axes[idx].bar(station_val[idx][0], station_val[idx][1], 
                         color=color_list[idx],
                         width=0.6, label=f'{idx+1}호선')

    # bar 상단 수치 
    axes[idx].bar_label(bar1, fmt ='{:,.0f}명', label_type='edge', padding=6, fontsize=18)

    axes[idx].set_title(f'2015년-2023년 {idx+1}호선 하차승객 TOP8 역사', fontsize=22, pad=15)
    axes[idx].set_ylim(bottom=0,top=max(first_tot) + 10**7*0.5*ylim_list[idx])

    axes[idx].set_xticks(list(range(len(station_val[idx][0]))), 
                         labels=station_val[idx][0],
                         fontsize=20)

    axes[idx].set_ylabel("명", fontsize=20, labelpad=10)
    axes[idx].legend(fontsize=20)
```
