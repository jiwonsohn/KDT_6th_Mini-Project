```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw = pd.read_csv('./tripadvisor_review.csv')
raw.head(5)
```

```python
raw.columns
```

```python
raw = raw.drop('User ID', axis=1)
raw.columns
```

```python
# 칼럼명 수정
cols = ['art_gallery', 'club','bar','restaurants', 'museum','resort',
        'park','beaches','theater','religeious']

raw.columns = cols
raw.columns
```

```python
raw.corr(numeric_only=True)
```

```python
after_DF = raw
after_DF.head(5)
```

```python
after_DF.corr()
```

#### 히트맵

```python
sns.heatmap(raw.corr(), cmap='viridis')
```

```python
plt.pcolor(after_DF.corr())
plt.xticks(np.arange(0.5, len(after_DF.columns), 1), after_DF.columns)
plt.yticks(np.arange(0.5, len(after_DF.columns), 1), after_DF.columns)
plt.title('Heatmap by plt.pcolor()', fontsize=20)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Month', fontsize=14)
plt.colorbar()
plt.show()
```
