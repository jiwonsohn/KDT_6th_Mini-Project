### 소비자물가 지수가 전기차 (대주제: 자동차 산업) 소비 패턴에 미치는 영향 분석

* 소비자 물가란? <br>
소비자물가지수는 각 가정이 생활을 위해 구입하는 상품과 서비스의 가격 변동을 알아보기 위해 작성하는통계<br>

* 정부가 경기를 판단하는 기초자료<br>
         * 일반적으로 물가는 경기가 상승국면에 있을 때는 수요 증가에 의하여 상승하는 경향이 있고, 하강 국면에서는 수요 감소에 의하여 하락하는 경향<br>
         * 물가가 높으면 경기가 좋은 상황 & 반대로 물가가 낮으면 경기가 나쁜 상황 이라 추정할 수 있음

* 화폐의 구매력 측정 수단 <br>
물가가 지속적으로 오르면 같은 돈으로 구입할 수 있는 상품의 양이 감소하게 되므로 돈의 가치 즉 화폐의 구매력은 떨어짐 <br>
         * 한정된 예산 안에서 최대한 합리적인 소비를 지향할 것으로 추정


출처: CPI물가지수, https://kostat.go.kr/ 

* 전기차는 경제적인가?<br>
              * 21년 충전요금 & 유류비 비교 --> 전기차(아이오닉5: 59원/km) vs 일반 내연기관(투싼: 가솔린:129원/km, 디젤:97원/km, 하이브리드:98원/km)<br>
                                          --> 전기차가 일반 내연기관 유지비용의 0.45배(가솔린 기준), 0.61배(디젤 기준), 0.61배(하이브리드 기준)<br>
                                               --> 평균 0.5배<br>
              * 적은 유지비용 --> 적은 부품 수 & 2-3배 긴 수명<br>
              * 세금혜택      --> 보유세는 내연기관 대비 약 50% 이상 저렴 / 통행료 및 공영 주차장 주차요금도 50% 할인 혜택<br>
              <br>

* 같은 화폐 단위에서 유류비, 유지비용, 세금혜택 각 방면에서 2배의 경제적 합리성을 가질 수 있음<br>

```python
import pandas as pd
import numpy as np
```

```python
# 물가지수 DF 로드 -> priceDB
price_file = r'C:\Users\KDP-43\Desktop\KDT_Pandas\pandas_project\소비자물가상승률_2013.2023.xlsx'
priceDB = pd.read_excel(price_file, skiprows=28, nrows=14)
priceDB
```

```python
# 0, 1번째 인덱스 행만 남겨두고 나머지 행은 삭제
priceDB.drop(priceDB.index[2:], axis=0, inplace=True)
```

```python
# 물가지수 DF 칼럼 'Unnamed: 0' -> '항목' 변경
print(f'물가 지수 DF 컬럼명: {priceDB.columns}')
print()
priceDB.rename(columns={priceDB.columns[0]:'항목'}, inplace=True)
priceDB
```

```python
# 물가지수 DF 인덱스를 칼럼 '항목'으로 재정의
priceDB.set_index(priceDB['항목'].values, inplace=True)
```

```python
# 물가지수 DF '항목' 칼럼 삭제
priceDB.drop("항목", axis=1, inplace=True)
```

```python
priceDB
```

```python
# 연도별 전기차 등록 추이 DF 로드
elec_car_file = r'C:\Users\KDP-43\Desktop\KDT_Pandas\pandas_project\전기차_연도별_추이.xlsx'
elec_car_DB = pd.read_excel(elec_car_file)
elec_car_DB
```

```python
# 홀수번째 인덱스 행 데이터는 전체 계 대비 해당 칼럼 시도 자동차 비율
# 짝수번째 인덱스행 데이터만 추출  
                                                 
need_range = []
for i in range(30):
    if not i%2: 
       need_range.append(i)

# 2013-2023 데이터 추출
elec_car_DB_2=elec_car_DB.iloc[need_range, :][1:12]
```

```python
elec_car_DB_2
```

```python
elec_car_DB_2.info()
```

```python
# 데이터프레임 데이터 타입 float64 -> int64로 변환
elec_car_DB_3 = elec_car_DB_2.astype(dtype='int64')
elec_car_DB_3
```

```python
# 데이터프레임 인덱스 변환
elec_car_DB_3.index = range(11)
elec_car_DB_3
```

```python
# 물가지수 DF과 칼럼 [2013, 2014, ..., 2023] 을 맞추기 위해 transpose()
elec_car_DB_4 = elec_car_DB_3.transpose()
elec_car_DB_4
```

```python
elec_car_DB_4.columns = elec_car_DB_3['연월']
elec_car_DB_4
```

```python
elec_car_DB_4.drop("연월", axis=0, inplace=True)
```

```python
# 최종 연도별 시도 단위 전기차 등록대수 DF w/ 칼럼 순서 역순 변환
elec_car_DB_5 = elec_car_DB_4.loc[:, ::-1]
elec_car_DB_5
```

```python
# 전국의 전기차 등록 소비 추이와 소비자물가를 비교하기 위해 마지막 '계' 인덱스 행만 남기기
elec_car_DB_5.drop(elec_car_DB_5.index[:-1], axis=0, inplace=True)
```

```python
# 물가지수 DF & 연도별 시도 단위 전기차 DF concat
price_eleccar_DB = pd.concat( [priceDB, elec_car_DB_5] )
price_eleccar_DB
```

```python
# 칼럼 확인
price_eleccar_DB.columns
```

```python
# price 칼럼을 elec_car_DB_5 칼럼 값으로 대체 ==> concat() 시 같은 칼럼 수 유지하기 위해
priceDB.columns = elec_car_DB_5.columns.values
priceDB.columns
```

```python
# 물가지수 DF & 연도별 시도 단위 전기차 DF concat
price_eleccar_DB = pd.concat( [priceDB, elec_car_DB_5] )
price_eleccar_DB
```

```python
# '계' 칼럼의 값들은 누적 값
# 따라서, '계' 칼럼 그대로 사용하면 실제 해당년도 전기차 등록 추이를 확인하기 위해서는 해당년도 계 - 작년도 계

# 연도별 전기차_등록_수 행 추가
elec_car = [1464.0]

for idx in range(2013,2023,1):
    cnt = (price_eleccar_DB.loc['계',idx+1] - price_eleccar_DB.loc['계',idx])
    elec_car.append( round(cnt,1) )

price_eleccar_DB.loc["전기차_등록_수"] = elec_car

# 전기차_등록_전년대비(%) 행 추가
elec_car_per = [np.nan]

for idx in range(2013,2023,1):
    per = (price_eleccar_DB.loc['전기차_등록_수',idx+1] - price_eleccar_DB.loc['전기차_등록_수',idx]) / price_eleccar_DB.loc['전기차_등록_수',idx] * 100
    elec_car_per.append( round(per,1) )


price_eleccar_DB.loc[ "전기차_등록_전년대비(%)"] = elec_car_per
```

```python
# 결과 DF
price_eleccar_DB
```

```python
# 소비자물가상승률 기준으로 결과 DF 오름차순 정렬
price_eleccar_DB.sort_values(by='소비자물가상승률(%)', axis=1)
```
