```python
# 각 컬럼별 결측치 여부 확인


# 수치형 음수(-1)
check_minus = {}
for col in raw_trainDF.columns[1:]:
    # print(f'===================={col}====================')
    # print( (raw_trainDF[col] < 0).sum())
    check_minus[col] = (raw_trainDF[col] < 0).sum()
print(check_minus)

# 수치형 0
check_zero = {}
for col in raw_trainDF.columns[1:]:
    check_zero[col] = (raw_trainDF[col] == 0).sum()
print(check_zero)

# 범주형 [Sex] value_counts()
print("\n[SEX value counts]")
print(raw_trainDF['Sex'].value_counts())
```

#### 다중회귀

##### 타겟/피쳐 분리
- 타겟: Rings
- 피쳐: Shell_weight,	'Height'

```python
# featureDF = drop_trainDF[['Shell_weight', 'Height']]
# featureDF = drop_trainDF[['Shell_weight', 'Diameter', 'Height', 'Shucked_weight']]

featureDF = drop_trainDF[['Shell_weight', 'Diameter', 'Height', 'Shucked_weight']]
targetSR = drop_trainDF['Rings']

print(f'featureDF: {featureDF.shape}, {featureDF.ndim}D')
print(f'targetSR: {targetSR.shape}, {targetSR.ndim}D')
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(featureDF, targetSR,
                                                    test_size=0.2,
                                                    random_state=12
                                                    )

print(f'X_train: {X_train.shape}, {X_train.ndim}D', f'y_train: {y_train.shape}, {y_train.ndim}D')
print(f'X_test: {X_test.shape}, {X_test.ndim}D', f'y_train: {y_test.shape}, {y_test.ndim}D')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
model = LinearRegression()

model.fit(X_train, y_train)

# model.score -> 내부에서 predict() 실행 ==> predict 결과로 R2 계수 산출 & 반환
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)
print()
print(f'train_score:	{train_score:.5f}\ntest_score:	{test_score:.5f}')				# r2_score 와 같은 수치

# 성능지표 => 오차계산 & 결정계수(r2) 계산
pre_target = model.predict(X_test)

# 손실/비용 계산 ==> 정답 & 예측값 / 0에 가까울수록 Good
print()
print(f'mse 	  : {mean_squared_error(y_test, pre_target):.5f}')
print(f'rmse 	  : {mean_squared_error(y_test, pre_target, squared=False):.5f}')
print(f'mae 	  : {mean_absolute_error(y_test, pre_target):.5f}')
print(f'r2 	  : {r2_score(y_test,pre_target):.5f}')
```

##### Cross_Validation

```python
from sklearn.model_selection import cross_val_score
cross_val_score(model, X_train, y_train, cv=5)
```

```python
from sklearn.model_selection import cross_val_predict
y_train_predict = cross_val_predict(model, X_train, y_train, cv=5)
y_test_predict = cross_val_predict(model, X_test, y_test, cv=5)
```
